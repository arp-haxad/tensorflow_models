## tf pracrise with different changes int he model

```python
import numpy as np
import matplotlib.pyplot as plt 
import tensorflow as tf
import seaborn as sns
sns.set()
```

```python
total_observation =1000
xs = np.random.uniform(-10,10,size = (total_observation,1))
zs = np.random.uniform(-10,10,size = (total_observation,1))
generated_inputs = np.column_stack((xs,zs))
noise = np.random.uniform(-1,1,(total_observation,1))
generated_outputs = 3*xs - 2*zs + 4 + noise
np.savez('tf_file2',inputs = generated_inputs,targets = generated_outputs)
```

## loading the data

```python
training_data = np.load('tf_file2.npz')
input_no = 2
output_no = 1
```

```python
training_data
```

```python
model  = tf.keras.Sequential([
                          tf.keras.layers.Dense(output_no,
                                           kernel_initializer=tf.random_uniform_initializer(minval=-0.1,maxval = 0.1),
                                           bias_initializer=tf.random_uniform_initializer(minval=-0.1,maxval = 0.1))
                            ])
            # setting learning rates by ourselves
            # op = tf.keras.optimizers.SGD(learning_rate = 0.2) 
custom_optimiser = tf.keras.optimizers.SGD(learning_rate = 0.02)               
model.compile(optimizer=custom_optimiser,loss = 'mean_squared_error')
model.fit(training_data['inputs'],training_data['targets'],epochs=100,verbose = 2)
```

```python
model.layers[0].get_weights()
weight_1 = model.layers[0].get_weights()[0]
bias_1 = model.layers[0].get_weights()[1]
weight_1
```
